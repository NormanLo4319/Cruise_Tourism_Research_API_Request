```python
# set up a parameters dictionary
params = {
    "location": '43.61, -116.21',
    "radius": 1000,
    "key": gkey
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
response = requests.get(base_url, params=params)
print(response)

# jsonify the response
address = response.json()
```

```python
# Extract the crews and passengers capacity of the ship

# Dependencies
import pandas as pd
import numpy as np

url = 'https://en.wikipedia.org/wiki/MS_Achille_Lauro'
```

```python
# Using pandas read.html to extract data from the html page
table = pd.read_html(url)
info = table[1]
info
```

```python
# Identify the class type for passenger capacity data
print(type(info[1][27]))

# Identify the crew capacity data
print(info[1][28])
```

```python
# Create a new colume for extracting the first element in each row
capacity = info[1].str.split(' ', n=1, expand=True)
info['new'] = capacity[0]
info
```

```python
# Create a for loop to save the capacity and crew data
passenger = []
crew = []

for i in info.index:
    if info[0][i] == "Capacity:":
#         print(info["new"][i])
# Need to replace commas with empty strings before coverting to interger
        passenger.append(int(info["new"][i].replace(',', '')))
    if info[0][i] == "Crew:":
#         print(info["new"][i])
# Need to replace commas with empty strings before coverting to interger
        crew.append(int(info["new"][i].replace(',', '')))

print(passenger)
print(crew)
```
